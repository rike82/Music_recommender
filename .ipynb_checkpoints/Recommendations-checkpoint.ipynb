```python
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id= config.client_id,
                                                           client_secret= config.client_secret))
```

```python
df_top100 = pd.read_csv('Top100_title_artist.csv')
df_top100['song_title'] = df_top100['song_title'].str.lower()

df_top100.head(50)
```

```python
def recommend(df):
    recommendation = random.choice(df['song_title'].values)
    artist_name = df.loc[df["song_title"] == recommendation]['artist'].iloc[0]
    return recommendation, artist_name
```

```python
# get id of the song
def get_song_id(user_input, i):

    user_json = sp.search(user_input, limit = 5)
    user_id = user_json["tracks"]["items"][i]["id"]
    user_name = user_json["tracks"]["items"][i]["name"] # Track name
#    except Exception as e:
#        print("Something went wrong, retrying...")
#        time.sleep(2)  # Kleine Pause vor erneutem Versuch

    return user_id, user_name
```

```python
def recommender():
    user_input = (input("Please enter one of your favourite songs!: "))
    i = 0
    song_id = get_song_id(user_input, i)[0]
    song_name = get_song_id(user_input, i)[1]
    
    
    display(play_song(song_id))
    yes_or_no = input("s this the song you mean? type y or n")
    while((yes_or_no == 'n')):
        i += 1
        if i >= 5: 
            print('Program starts again... choose another song!')
            return recommender() 
        song_id, song_name = get_song_id(user_input, i)  # Hol ID und Namen
              # Nur ID an play_song übergeben
        display(play_song(song_id))
        yes_or_no = input("s this the song you mean? type y or n")
        
    
    if song_name.lower() in df_top100['song_title'].values:
        recommendation = recommend(df_top100)
    
        print('Song is in the top 100!')
        print(f"Here is a recommanded song for you: '{recommendation[0]}' by '{recommendation[1]}' ")
        song_name_100 = recommendation[0]
        song_id_top100 = get_song_id(song_name_100, i)[0]
        display(play_song(song_id_top100))
        
    else:
        print('Song not found in top100!')
        print('This might be another song you like: ')
        find_alternative(song_id)
            
```

```python
def load(filename): 
    try: 
        with open(filename, "rb") as f: 
            return pickle.load(f) 
        
    except FileNotFoundError: 
        print("File not found!") 
```

```python
def find_alternative(song_id):
    scaler= load("scaler.pickle")
    kmeans_model = load("model.pickle")
    audiofeatures = sp.audio_features(song_id)
    df_song = pd.DataFrame(audiofeatures)
    columns_scaled = ['danceability', 'energy', 'loudness','speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']
    df_song[columns_scaled] = scaler.transform(df_song[columns_scaled])
    df_song['KMeans_Cluster'] = kmeans_model.predict(df_song[columns_scaled])
    predicted_cluster = df_song['KMeans_Cluster'].iloc[0]
    model_df = pd.read_csv('Train_Playlist.csv')
    df_sample = model_df[model_df['KMeans_Cluster'] == predicted_cluster]
    song = df_sample.sample()
    new_song_id = song['id'].iloc[0]
    display(play_song(new_song_id))


```

```python
def play_song(track_id):
    return IFrame(src="https://open.spotify.com/embed/track/"+track_id,
       width="320",
       height="80",
       frameborder="0",
       allowtransparency="true",
       allow="encrypted-media",
      )
```
