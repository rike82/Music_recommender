```python
# K-means
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn import cluster, datasets
from sklearn.preprocessing import StandardScaler
from matplotlib.lines import Line2D
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv('Playlist.csv', index_col=[0])
df
```

```python
my_scaler = StandardScaler()
columns_scaled = ['danceability', 'energy', 'loudness','speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']
df[columns_scaled] = my_scaler.fit_transform(df[columns_scaled])
df
```

```python
# Apply K-Means clustering
kmeans_rike = KMeans(n_clusters = 20, random_state = 42)
df['KMeans_Cluster'] = kmeans_rike.fit_predict(df[columns_scaled])
df.to_csv('Train_Playlist.csv', index = False)
df
```

```python
labels = kmeans_rike.labels_
labels
```

```python
np.unique(labels)
```

```python

pd.Series(df['KMeans_Cluster']).value_counts().sort_index()
```

```python
Sum_of_squared_distances = []
K = range(3,20) 
for num_clusters in K :
 kmeans = KMeans(n_clusters=num_clusters)
 kmeans.fit(df[columns_scaled])
 Sum_of_squared_distances.append(kmeans.inertia_)
 
plt.plot(K,Sum_of_squared_distances,'go-')
plt.xlabel('Values of K') 
plt.ylabel('Sum of squared distances/Inertia') 
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
K = range(3,20) 
silhouette = []


for num_clusters in K :
    kmeans = KMeans(n_clusters=num_clusters)
    kmeans.fit(df[columns_scaled])

    silhouette.append(silhouette_score(df[columns_scaled], kmeans.predict(df[columns_scaled])))
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, silhouette, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette score')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Silhouette Method showing the optimal k')
```
